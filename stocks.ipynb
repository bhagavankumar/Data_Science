```python
import pandas as pd
stocks=pd.read_csv("TSX 60.csv")
```

```python
stocks.head()
```

```python
stocks.drop(stocks.columns[stocks.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
```

```python
stocks.head()
```

```python
stocks.tail()
```

```python
stocks.dropna(subset=['STOCKS'], how='all', inplace=True)
```

```python
stocks.tail()
```

```python
def percentage_change(col1,col2):
    return ((col2 - col1) / col1) * 100
```

```python
stocks['a'] = percentage_change(stocks['1-Jul-24'], stocks['13-Jul-24'])    
stocks['b'] =  percentage_change(stocks['13-Jul-24'], stocks['27-Jul-24'])
stocks['c'] =  percentage_change(stocks['27-Jul-24'], stocks['3-Aug-24'])
stocks['d'] =  percentage_change(stocks['3-Aug-24'], stocks['10-Aug-24'])
stocks['e'] =  percentage_change(stocks['10-Aug-24'], stocks['17-Aug-24'])
stocks['f'] =  percentage_change(stocks['17-Aug-24'], stocks['24-Aug-24'])
stocks['g'] =  percentage_change(stocks['24-Aug-24'], stocks['31-Aug-24'])
stocks.head()
```

```python
from matplotlib import pyplot as plt
plt.scatter(stocks['a'],stocks['b'])
plt.title('stocks percentage change within 2 weeks')
plt.xlabel('1st week changes')
plt.ylabel('2nd week changes')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
x=stocks['a'].values.reshape(-1,1)
y=stocks['b'].values.reshape(-1,1)
regressor=LinearRegression()
regressor.fit(x,y)
print(regressor.coef_)
print(regressor.intercept_)
```

```python
regline=regressor.predict(x)
plt.scatter(stocks['a'],stocks['b'],color='blue')
plt.plot(sorted(stocks['a'].tolist()),regline,'r')
plt.title('stocks percentage change within 2 weeks')
plt.xlabel('1st week')
plt.ylabel('2nd week')
plt.show()
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsRegressor
first_week=stocks['a']
second_week=stocks['b'].tolist()
x=np.array(first_week).reshape(-1,1)
y=np.array(second_week)
```

```python
knnregressor=KNeighborsRegressor(n_neighbors=15)
new_percent_change=0.5
knnregressor.fit(x,y)
print(knnregressor.predict(np.array([new_percent_change]).reshape(-1,1)))
```

```python
#using classificaion instead of regression
themedian=np.median(stocks['g'])
stocks['abovemedianstocks']=1*(stocks['g']>themedian)
stocks.head()
```

```python
from sklearn.model_selection import train_test_split
x=stocks[['a','b','c','d','e','f']]
y=np.array(stocks['abovemedianstocks'])
trainingx,testx,trainingy,testy=train_test_split(x,y,random_state=1)
from sklearn.neighbors import KNeighborsClassifier
knnclassifier=KNeighborsClassifier(n_neighbors=15)
trainingx,testx,trainingy,testy=train_test_split(x,y,random_state=1)
knnclassifier.fit(trainingx,trainingy)
predicted=knnclassifier.predict(testx)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(testy,predicted))
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
precision=precision_score(testy,predicted)
recall=recall_score(testy,predicted)
```

```python
print(precision)
print(recall)
```
